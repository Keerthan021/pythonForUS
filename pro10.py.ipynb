```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/home/mcacnj/DAP Datasets/AirPassengers.csv')
df.head()
```

```python
df.columns=['Date','Number of Passenger']
df.head()
```

```python
def plot_df(df, x, y, title="", xlabel='Date', ylabel='Number of Passenger', dpi=100):
    plt.figure(figsize=(15,4), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()
plot_df(df, x=df['Date'], y=df['Number of Passenger'], title='Number of US Airline Passengers from 1949 to 1960')
```

```python
x = df['Date'].values
y1 = df['Number of Passenger'].values

fig, ax = plt.subplots(1, 1, figsize=(16, 5), dpi=120)
plt.fill_between(x, y1=y1, y2=0, alpha=0.5, linewidth=2, color='seagreen')
plt.ylim(-800, 800)
plt.title('Air passenger (Two side view)', fontsize=16)
plt.hlines(y=0, xmin=np.min(df['Date']), xmax=np.max(df['Date']), linewidth=5)
plt.show()
```
