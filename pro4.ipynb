```python
import pandas as pd
import io
import numpy as np

data = pd.read_csv("/home/mca/Dataset/file1.csv")
data.head(10)
```

```python
# cleaning the data

data.drop_duplicates(subset = ['name'],keep="first",inplace = True)
data.head(10)
```

```python
# To count empty field space

data.isnull().sum()

#missing data
print(data.shape)
```

```python
#dropping the missing rows
df_dropped = data.dropna()
data.head(10)
```

```python
#fill with provided value(Hobbies)
data['Hobbies'] = data['Hobbies'].fillna('Carrom')
data.head(10)
```

```python
# fill with provided value(mobile number)

mean_val = data['Mobile Number'].mean()
data['Mobile Number'] = data['Mobile Number'].fillna(mean_val)
data.head(10)
```
